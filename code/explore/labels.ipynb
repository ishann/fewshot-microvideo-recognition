```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Read in the data
DATA_DIR = '../../3MASSIV/data/labels'
trn_labels = pd.read_csv(os.path.join(DATA_DIR, '3massiv_train.csv'))
trn_concepts = pd.read_csv(os.path.join(DATA_DIR, '3massiv_concept_train.csv'))
val_labels = pd.read_csv(os.path.join(DATA_DIR, '3massiv_val.csv'))
val_concepts = pd.read_csv(os.path.join(DATA_DIR, '3massiv_concept_val.csv'))

TOPICS = ["Dance", "Couple Romance", "Heartbreak Romance", "Memes", "Pranks"]
VIDEO_FORMATS = ["animation", "image", "self-shot", "special-effects", "videos-with-text"]
LANGUAGES = ["English", "Hindi", "Haryanvi", "Bengali", "Tamil"]
```

```python
print("Train labels:\n", trn_labels.head())
print("\nVal labels:\n", val_labels.head())
print("\nTrain concepts:\n", trn_concepts.head())
print("\nVal concepts:\n", val_concepts.head())
```

```python
# Plot histogram of train labels sorted in descending order.
trn_labels["Topic"].value_counts(sort=True, ascending=True).plot.barh(title="Topic")
```

```python
trn_labels["VideoFormat"].value_counts(sort=True, ascending=True).plot.barh(title="VideoFormat")
```

```python
trn_labels["Language"].value_counts(sort=True, ascending=True).plot.barh(title="Language")
```

```python
trn_concepts["concept"].value_counts(sort=True, ascending=True).plot.barh(title="Concept")
```

```python
labels = pd.concat([trn_labels, val_labels])
concepts = pd.concat([trn_concepts, val_concepts])
```

```python
labels_f_topic = labels[labels["Topic"].isin(TOPICS)]
labels_f_VFormat = labels_f_topic[labels_f_topic["VideoFormat"].isin(VIDEO_FORMATS)]
labels_f_Language = labels_f_VFormat[labels_f_VFormat["Language"].isin(LANGUAGES)]

f_labels = labels_f_Language
```
